```python
import time
data_atual = time.localtime()
ano, mes_, dia, hora, minuto, segundo, diadasemana_, diadoano_,isdst = data_atual

import matplotlib.pyplot as plt
import numpy as np
from collections import Counter



def descobrir_mes(numero_do_mes):
    
    for l, each in enumerate(dicionario_meses):

        if numero_do_mes <= 11:
            if each == numero_do_mes:
                mes_correspondente = dicionario_meses[l]
                return mes_correspondente + f'/{ano - 2000}'
            else:
                pass

        if numero_do_mes > 11:
            if each == numero_do_mes - 11:
                mes_correspondente = dicionario_meses[l]
                return mes_correspondente + f'/{(ano + 1) - 2000}'
            else:
                pass

        else:
            pass

    

dicionario_meses = {
    0 : "janeiro",
    1 : "fevereiro",
    2 : "março",
    3 : "abril",
    4 : "maio",
    5 : "junho",
    6 : "julho",
    7 : "agosto",
    8 : "setembro",
    9 : "outubro",
    10 : "novembro",
    11 : "dezembro",
}


meses_randomificados = 22
vendas_loja1 = np.random.randint(5000, 9000, meses_randomificados)
vendas_loja2 = np.random.randint(5000, 9000, meses_randomificados)
meses = np.arange(1, meses_randomificados + 1)

#Top3 de todas as lojas
vendas_lojasfull = list(zip(vendas_loja1, vendas_loja2))
medias = []
for l, each in enumerate(vendas_lojasfull):

    maior_valor = max(vendas_lojasfull[l])
    menor_valor = min(vendas_lojasfull[l])
    media = (maior_valor + menor_valor) / len(vendas_lojasfull[l])
    medias.append(media)


dicio_lojas_geral = dict(zip(meses, medias))
top3_geral = Counter(dicio_lojas_geral)
top3_geral = top3_geral.most_common(3)


print(f'\033[42m\033[1m{"Top3 das médias das lojas":^45}\033[0m')
print('=' * 45)
print(f'|{"Melhor Mês":^17}|{"Maior média de venda":^25}|')
print('=' * 45)
for l, each in enumerate(top3_geral):
    mês, media_mes = each
    
    print('|{}º{:^15}|{:^25}|'.format(1 + l, descobrir_mes(mês), media_mes))
print('=' * 45)
print('\n')



#Top3 Loja 1
dicio_loja1 = dict(zip(meses, vendas_loja1))
top3_loja1 = Counter(dicio_loja1)
top3_loja1 = top3_loja1.most_common(3)

print(f'\033[41m{"Top3 Loja 1":^45}\033[0m')
print('=' * 45)
print(f'|{"Melhor Mês":^17}|{"Maior média ":^25}|')
print('=' * 45)

for l, each in enumerate(top3_loja1):    
    mês, alcance = each
    print('|{}º{:^15}|{:^25}|'.format(1 + l, descobrir_mes(mês), alcance))
print('=' * 45)
print('\n')



#Top3 Lojas 2
dicio_loja2 = dict(zip(meses, vendas_loja2))
top3_loja2 = Counter(dicio_loja2)
top3_loja2 = top3_loja2.most_common(3)

print(f'\033[44m{"Top3 Loja 2":^45}\033[0m')
print('=' * 45)
print(f'|{"Melhor Mês":^17}|{"Maior média ":^25}|')
print('=' * 45)

for l, each in enumerate(top3_loja2):    
    mês, alcance = each
    print('|{}º{:^15}|{:^25}|'.format(1 + l, descobrir_mes(mês), alcance))
print('=' * 45)
print('\n')



##########################  GRÁFICO  #############################
#figure 1
plt.figure(1, figsize=(15, 5), facecolor = '0.5')
plt.suptitle('Comparativo gráfico de resultados da LOJA 1', fontsize=20, color = 'white')

plt.subplot(1, 2, 1)
plt.bar(meses, vendas_loja1, color = 'red')
plt.xlabel("Meses")
plt.ylabel("Vendas")
plt.axis([0, len(meses), min(vendas_loja1) - 500, max(vendas_loja1) + 200])

plt.subplot(1, 2, 2)
plt.plot(meses, vendas_loja1, 'r--.')
plt.xlabel("Meses")
plt.ylabel("Vendas")
plt.axis([0, len(meses), min(vendas_loja1) - 500, max(vendas_loja1) + 200])
plt.show()



#figure 2
plt.figure(2, figsize=(15, 5), facecolor = '0.5')
plt.suptitle('Comparativo gráfico de resultados da LOJA 2', fontsize=20, color = 'white')

plt.subplot(1, 2, 1)
plt.bar(meses, vendas_loja2)
plt.xlabel("Meses")
plt.ylabel("Vendas")
plt.axis([0, len(meses), min(vendas_loja2) - 500, max(vendas_loja2) + 300])

plt.subplot(1, 2, 2)
plt.plot(meses, vendas_loja2, 'c--.')
plt.xlabel("Meses")
plt.ylabel("Vendas")
plt.axis([0, len(meses), min(vendas_loja2) - 500, max(vendas_loja2) + 300])
plt.show()


```
